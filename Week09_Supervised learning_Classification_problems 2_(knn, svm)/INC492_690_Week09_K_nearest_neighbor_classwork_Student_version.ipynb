#### KNN algorithm

```python
# Create knn function
def k_nearest_neighbors(data, predict, k=2):
    distances = []
    for group in data:
        # Create variable for collecting the distances
        for feature in data[group]:
          # Calucalate distance between data to new data
            euclidean_distance = np.linalg.norm(np.array(feature)-np.array(predict))
            distances.append([euclidean_distance, group])

# Evaluate the top k data that near to new data
    votes = [i[1] for i in sorted(distances)[:k]]
    vote_result = Counter(votes).most_common(1)[0][0]
    confidence = Counter(votes).most_common(1)[0][1] / k
    return vote_result, confidence

```

#### Let us predict for new features and visualize

```python
# Add new data point
new_features = [6,6]

# Predicted group of new data point
result = k_nearest_neighbors(dataset, new_features, k=2)
result
```
